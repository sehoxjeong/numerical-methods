## Joint Kernel Density Estimate

```python
# Load the `penguins` dataset.
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.species.unique()
```

```python
# Show the joint distribution using kernel density estimation.
kde = sns.jointplot(data=penguins,
                    x='bill_length_mm', y='bill_depth_mm', hue='species',
                    kind='kde',
                    cmap='Reds')
```

## Three-Way ANOVA

```python
# Load the example exercise dataset.
exercise = sns.load_dataset('exercise')
```

```python
exercise.head()
```

```python
# Draw a pointplot to show pulse as a function of three categorial factors.
sns.set_theme(style='whitegrid')
twanova = sns.catplot(data=exercise,
                      x='time', y='pulse', hue='kind', col='diet',
                      capsize=.1, palette='YlGnBu_d', errorbar='se', kind='point', height=6, aspect=0.75)
twanova.despine(left=True)
```
